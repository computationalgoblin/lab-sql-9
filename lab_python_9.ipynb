Now that i had set up the database i will create dataframe with the number of rentals for each customer in may

```python
rentals_for_customer_may = pd.read_sql_query('select count(rental_id) as counter, customer_id from rentals_may group by customer_id order by customer_id', engine)
rentals_for_customer_may
```

Lets do the same with june table

```python
rentals_for_customer_june = pd.read_sql_query('select count(rental_id) as counter, customer_id from rentals_june group by customer_id order by customer_id', engine)
rentals_for_customer_june
```

Now let's write a function that checks if a given customer borrowed more or less films in the month of June as compared to May

```python
final_df = pd.merge(rentals_for_customer_may, rentals_for_customer_june, on = "customer_id", suffixes= ("_may", "_june"))
final_df
```

```python

def more_rentals(i):
    if i['counter_may'] > i['counter_june']:
        return 'less'
    elif i['counter_may'] < i['counter_june']:
        return 'more'
    else:
        return 'same'

final_df['solution'] = final_df.apply(more_rentals, axis=1)
final_df
```
